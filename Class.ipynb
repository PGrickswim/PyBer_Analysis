```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_axis = np.arange(0, 6, 0.1)
x_axis
```

```python
sin = np.sin(x_axis)
sin
```

```python
cos = np.cos(x_axis)
cos
```

```python
plt.plot(x_axis, sin)
plt.plot(x_axis, cos)

plt.show()
```

```python
for x in x_axis:
    np.exp(x)
```

```python
e_x = [np.exp(x) for x in x_axis]
e_x
```

```python
plt.plot(e_x)
plt.show()
```
